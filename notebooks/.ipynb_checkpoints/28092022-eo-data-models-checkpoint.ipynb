# Data Modelling

## Models

We chose to use three main classical models to compare to Neural Network Models:
- SARIMA (with Auto-ARIMA)
- SARIMA (with Gridsearch)
- Random Forest (baseline)
- Random Forest (with gridsearch)

The NN model of choice was the __LSTM__ neural network due to its application to our type of data.

## Procedures
The next segment will show a step-by-step approach for each decison taken during the data modeling process.

### Imports

Due to the extensive library imports required, their explanation will be done in the code's comments:

```python
# Data manipulation
# ==============================================================================
import numpy as np
import pandas as pd
import os

# Plots
# ==============================================================================
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
plt.rcParams['lines.linewidth'] = 1.5
%matplotlib inline

# Modeling and Forecasting
# ==============================================================================
from pmdarima.arima import auto_arima
from statsmodels.tsa.statespace.sarimax import SARIMAX
from skforecast.ForecasterAutoreg import ForecasterAutoreg
from sklearn.ensemble import RandomForestRegressor
from tensorflow.keras.layers import LSTM
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Hyperparamter Optmization
# ==============================================================================
from skforecast.model_selection_statsmodels import grid_search_sarimax
from skforecast.model_selection import grid_search_forecaster
from tensorflow.keras.callbacks import EarlyStopping

# Data Preprocess
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
from sklearn.preprocessing import MinMaxScaler

# Model Metrics
# ==============================================================================
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

# Warnings configuration
# ==============================================================================
import warnings
warnings.filterwarnings('ignore')
```

Instantiating the errors' list that we'll use for each model:

```python
rsmes = []
maes = []
r2s = []
```

Loading the datasets:

```python
MONTHLY_DATA_PATH = os.path.abspath(r"..\data").replace('\\','/') + '/processed/monthly_rainfall.xlsx'
WEEKLY_DATA_PATH = os.path.abspath(r"..\data").replace('\\','/') + '/processed/weekly_rainfall_avg.xlsx'
DAILY_DATA_PATH = os.path.abspath(r"..\data").replace('\\','/') + '/processed/daily_rainfall.xlsx'

monthly_df = pd.read_excel(MONTHLY_DATA_PATH, 
                           index_col=[0],
                           parse_dates=[0])

monthly_df.index.freq = 'MS'

weekly_df = pd.read_excel(WEEKLY_DATA_PATH,
                          index_col=[0],
                          parse_dates=[0])

daily_df = pd.read_excel(DAILY_DATA_PATH,
                         index_col=[0], 
                         parse_dates=[0])
```

Checking data integrity:

```python
monthly_df.head()
```

```python
monthly_df.shape
```

Data shape looks fine. As explained in the [previous notebook](22092022-vtm-decomposing-time-series-and-sazonality.ipynb), we opted to use the monthly average dataset, since it's lower noise will make the predictions more accurate, since we'll not be transfering over the inherent residual error from the data:

```python
monthly_df.plot(figsize=(10,5))
plt.show()
```

```python
data = monthly_df.copy()
```

### Spliting the data

The train-test split from the data will use 24 steps - in other words, 24 months, making up the total of 2 whole years of that. This will give us a 75/25 train test split, which is great given the dataset's size.

```python
steps = 24
data_train = data[:-steps]
data_test  = data[-steps:]

print(f"Train dates : {data_train.index.min()} --- {data_train.index.max()}  (n={len(data_train)})")
print(f"Test dates  : {data_test.index.min()} --- {data_test.index.max()}  (n={len(data_test)})")

fig, ax = plt.subplots(figsize=(12,6))
data_train['monhtly_avg'].plot(ax=ax, label='train')
data_test['monhtly_avg'].plot(ax=ax, label='test')
ax.legend();
```

### SARIMA with Auto ARIMA model:

```python
stepwise_model = auto_arima(data_train, max_order=100, start_p=0, max_p = 4, start_q = 0, max_q = 3, m = 12,
                  seasonal = True, start_P = 0, max_P = 3, start_Q = 0, max_Q = 3,
                  trace = True, error_action = 'ignore', suppress_warnings = True,
                  stepwise = False, information_criterion='aic' )

stepwise_model.fit(data_train)
predictions = stepwise_model.predict(n_periods = steps)
```

```python
# Plot
# ==============================================================================
fig, ax = plt.subplots(figsize=(12,6))
data_train['monhtly_avg'].plot(ax=ax, label='train')
data_test['monhtly_avg'].plot(ax=ax, label='test')
predictions.plot(ax=ax, label='predictions')
ax.legend();

error_mse = mean_squared_error(
                y_true = data_test,
                y_pred = predictions
            )

error_mae = mean_absolute_error(
                y_true = data_test,
                y_pred = predictions
            )

rsmes.append(error_mse**0.5)

maes.append(error_mae)

print(f"Mean Absolute Error (MAE): {error_mae:.2f}")

print(f"Root Mean Squared Error (RSME): {error_mse**0.5:.2f}")

print(f"Correlation Coefficient: {r2_score(data_test, predictions):.2f}")
r2s.append(r2_score(data_test, predictions))

# Plotting the predicted data versus the real data
fig = plt.figure(figsize=(8,8))
l = plt.plot(predictions.values, data_test.values, 'bo')
plt.setp(l, markersize=10)
plt.setp(l, markerfacecolor='C0')
plt.title('Predicted Value x True Value Comparison', fontsize=12)
plt.ylabel("True Value", fontsize=12)
plt.xlabel("Predict Value", fontsize=12)

# Shows the predicted and original data
xl = np.arange(min(data_test.values), 1.2*max(data_test.values),(max(data_test.values)-min(data_test.values))/10)
yl = xl
plt.plot(xl, yl, 'r--')
plt.show()
```

### Gridsearch optimized SARIMA

```python
list_order = [(p, d, q) for p in range(1) for d in range(3) for q in range(4)]
list_order_s = [(P, D, Q, 12) for P in range(3) for D in range(3) for Q in range(4)]

param_grid = {'order': list_order,
             'seasonal_order': list_order_s}

results_grid = grid_search_sarimax(
                y = data_train,
                param_grid = param_grid,
                initial_train_size = len(data_train) - steps,
                fixed_train_size = True,
                steps = 1,
                metric = 'mean_squared_error',
                refit = False,
                verbose = False)
results_grid
```

```python
order = results_grid.iloc[0,2]
order_s = results_grid.iloc[0,3]
trend = results_grid.iloc[0,4]

modelo = SARIMAX(data_train, order = order, seasonal_order= order_s, trend = trend)

results = modelo.fit()

predictions = results.predict(start = len(data_train), end=len(data_train)+len(data_test)-1)
```

```python
# Plot
# ==============================================================================
fig, ax = plt.subplots(figsize=(12,6))
data_train['monhtly_avg'].plot(ax=ax, label='train')
data_test['monhtly_avg'].plot(ax=ax, label='test')
predictions.plot(ax=ax, label='predictions')
ax.legend();

error_mse = mean_squared_error(
                y_true = data_test,
                y_pred = predictions
            )

error_mae = mean_absolute_error(
                y_true = data_test,
                y_pred = predictions
            )

rsmes.append(error_mse**0.5)

maes.append(error_mae)

print(f"Mean Absolute Error (MAE): {error_mae:.2f}")

print(f"Root Mean Squared Error (RSME): {error_mse**0.5:.2f}")

print(f"Correlation Coefficient: {r2_score(data_test, predictions):.2f}")
r2s.append(r2_score(data_test, predictions))

# Plotting the predicted data versus the real data
fig = plt.figure(figsize=(8,8))
l = plt.plot(predictions.values, data_test.values, 'bo')
plt.setp(l, markersize=10)
plt.setp(l, markerfacecolor='C0')
plt.title('Predicted Value x True Value Comparison', fontsize=12)
plt.ylabel("True Value", fontsize=12)
plt.xlabel("Predict Value", fontsize=12)

# Shows the predicted and original data
xl = np.arange(min(data_test.values), 1.2*max(data_test.values),(max(data_test.values)-min(data_test.values))/10)
yl = xl
plt.plot(xl, yl, 'r--')
plt.show()
```

### LSTM Neural Network

```python
scaler = MinMaxScaler()
scaled_train = scaler.fit_transform(data_train)
scaled_test = scaler.transform(data_test)
```

```python
n_input = 12
n_features = 1

generator = TimeseriesGenerator(scaled_train, 
                                scaled_train, 
                                length = n_input,
                                batch_size = 3)
```

```python
X, y = generator[0]
```

```python
# Understanding batch separation and expected predicition
print('Expected data by batch and prediction')
print(f'First batch: \n{X.flatten()}');
print(f'Prediction: \n {y}');
```

```python
model = Sequential([
            LSTM(100, activation='relu', input_shape=(n_input, n_features)),
            Dense(1, activation = 'sigmoid')])
model.compile(optimizer='adam', loss='mse')
model.fit(generator, epochs = 50)

test_predictions = []

first_eval_batch = scaled_train[-n_input:]
current_batch = first_eval_batch.reshape((1, n_input, n_features))

for i in range(len(data_test)):
    # gets time 1 forecast beforehand (index [0] is only to get the number instead of the [array])
    current_pred = model.predict(current_batch)[0]
    
    # prediction
    test_predictions.append(current_pred) 
    
    # update the round to include the forecast and discard the first value
    current_batch = np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```

```python
predictions = scaler.inverse_transform(test_predictions)
predictions = [prediction[0] for prediction in predictions]
predictions = pd.DataFrame(predictions, index = data_test.index, columns = ['monhtly_avg'])


# Plot
# ==============================================================================
fig, ax = plt.subplots(figsize=(12,6))
data_train['monhtly_avg'].plot(ax=ax, label='train')
data_test['monhtly_avg'].plot(ax=ax, label='test')
predictions.plot(ax=ax, label='predictions')
ax.legend();


error_mse = mean_squared_error(
                y_true = data_test,
                y_pred = predictions
            )

error_mae = mean_absolute_error(
                y_true = data_test,
                y_pred = predictions
            )

rsmes.append(error_mse**0.5)

maes.append(error_mae)

print(f"Mean Absolute Error (MAE): {error_mae:.2f}")

print(f"Root Mean Squared Error (RSME): {error_mse**0.5:.2f}")

print(f"Correlation Coefficient: {r2_score(data_test, predictions):.2f}")
r2s.append(r2_score(data_test, predictions))

# Plotting the predicted data versus the real data
fig = plt.figure(figsize=(8,8))
l = plt.plot(predictions.values, data_test.values, 'bo')
plt.setp(l, markersize=10)
plt.setp(l, markerfacecolor='C0')
plt.title('Predicted Value x True Value Comparison', fontsize=12)
plt.ylabel("True Value", fontsize=12)
plt.xlabel("Predict Value", fontsize=12)

# Shows the predicted and original data
xl = np.arange(min(data_test.values), 1.2*max(data_test.values),(max(data_test.values)-min(data_test.values))/10)
yl = xl
plt.plot(xl, yl, 'r--')
plt.show()
```

### Ranking each model

Finally, we can rank each model's performance in a more clear and concise way. The results are as follows:

```python
rank = pd.DataFrame({
    'Model': ['SARIMA (AIC)', 
              'SARIMA (GridSearch)', 
              'Random Forest (GridSearch)',
              'Long short-term memory'],
    'RSME': [f'{rsme:.2f}' for rsme in rsmes],
    'MAE': [f'{mae:.2f}' for mae in maes],
    'r ': [f'{r2:.4f}' for r2 in r2s]})
rank.sort_values(by='RSME', ascending = True).set_index('Model')
```

## Conclusions

### Classical Model's Performance

Although SARIMA would be the first choice for such a dataset, its performace was below expectations, even with the Gridsearch Optmiziation implemented. Surprisingly, Random Forest (with gridsearch implemented) was the best model within the classical models category, coming very close to the neutal network's score.

### LSTM Model's Performance

It was expected that LSTM would yield the best results, however, they were marginally better than the random forest model. Perhaps the size od the dataset could influence the outcome; during the data exploration, it was hypothesized that by adding more rainfall data, the neural networks could perform better. This is quite a difficult conumdrum, since the model can only be updated after the end of the full calendar year's rainfall measurements, rendering the updating model slow and making its application more focused in monitoring and making short-term forecasts rather than predicting in a month-by-month basis.

### Final considerations

Nonetheless, the application was overall a success from the team's viewpoint. Considering the dataset's scale, size, and that it wasn't a classical, true-and-tested dataset such as the Titatinic, Iris, or Tips dataset, the preditions were accurate enough to justify a continuous moniroting and study of the area, perhaps adjusting the data collection system to a telemetry-based, real-time approach, instead of a daily collection with and a monthly average tabulation.
